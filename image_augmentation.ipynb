```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
from PIL import Image
from tensorflow.keras import Sequential, models, layers
```

```python
datagen = ImageDataGenerator(
    zoom_range = 0.03,
    rotation_range = 20,
    horizontal_flip = True,
    vertical_flip = True,
    brightness_range = [0,0.3]
)
```

```python
#list of paths of all of the images
def img_list(path):
    directory_list = []
    for i in os.listdir(path):
        full_path = os.path.join(path, i)
        directory_list.append(full_path)
    return directory_list
```

```python
einstein_path = 'einstein_2nd'
elon_path = 'elon musk'
ronaldinho_path = 'ronaldinho'
```

```python
ronaldinho_directories = img_list(ronaldinho_path)
einstein_directories = img_list(einstein_path)
elon_directories = img_list(elon_path)
```

```python
def augmentation_images(image_paths, saving_directory):
    image_list = []
    for i in image_paths:
        img = cv2.imread(i)
        if img is None:
            continue
        resized_images = cv2.resize(img, (255,255))
        reshaped_images = resized_images.reshape((1,) + resized_images.shape)
        image_list.append(reshaped_images)
        
    for i in image_list:    
        count = 0
        for batch in datagen.flow(i, batch_size=1, save_to_dir= saving_directory, save_prefix = 'augmented'):
            count +=1
            if count > 3:
                break
```

```python
einstein_augmentation = augmentation_images(einstein_directories, einstein_path)
ronaldinho_augmentation = augmentation_images(ronaldinho_directories, ronaldinho_path)
elon_augmentation = augmentation_images(elon_directories, elon_path)
```
